```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.005, 0.995, 200)

def entropy(p):
    q = (1 - p)
    H = -(p * np.log2(p) + q * np.log2(q))
    return H

y = entropy(x)

plt.rcParams['figure.dpi'] = 300
plt.plot(x, y)
plt.title("Coin toss entropy")
plt.xlabel("P(heads)")
plt.ylabel("Entropy")
```

```python
def loss(a):
    return -np.log2(a)

x = np.linspace(0.005, 1.0, 201)

y = loss(x)

plt.plot(x, y)
plt.title("Loss")
plt.xlabel("a")
plt.ylabel("Loss(a)")
y[-1]
```
